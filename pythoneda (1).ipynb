```python
plt.figure(figsize=(10,5))
sales_product = df.groupby('PRODUCTLINE', as_index=False)['SALES'].sum().sort_values(by='SALES', ascending=False)
ax = sns.barplot(x='PRODUCTLINE', y='SALES', data=sales_product)
ax.bar_label(ax.containers[0])
plt.title("Total Sales by Product Line")
plt.show()
```

Classic Cars dominate sales by a large margin.  
- Other strong performers: Vintage Cars and Planes.  

```python
plt.figure(figsize=(9,5))
sales_country = df.groupby('COUNTRY', as_index=False)['SALES'].sum().sort_values(by='SALES', ascending=False).head(10)
ax = sns.barplot(x='COUNTRY', y='SALES', data=sales_country)
ax.bar_label(ax.containers[0])
plt.title("Top 10 Countries by Sales")
plt.xticks(rotation=45)
plt.show()
```

USA Leads the sales chart by a huge difference. 

```python
# convert ORDERDATE to datetime
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
# Ensure ORDERDATE is datetime
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Create a Year-Month column
df['YearMonth'] = df['ORDERDATE'].dt.to_period('M').astype(str)

# Group by YearMonth
sales_month = df.groupby('YearMonth', as_index=False)['SALES'].sum()

# Plot
plt.figure(figsize=(12,5))
ax = sns.barplot(x='YearMonth', y='SALES', data=sales_month)
ax.bar_label(ax.containers[0])
plt.title("Monthly Sales Trend")
plt.show()
```

Sales peak in November–December,indicating holiday season demand.  
A noticeable dip in mid-year months suggests off-season trends.  

```python
plt.figure(figsize=(8,5))
ax = sns.scatterplot(x='QUANTITYORDERED', y='SALES', data=df, alpha=0.6)
plt.title("Quantity vs Sales")
plt.show()
```

As the quantity ordered increases, sales generally rise, showing a positive correlation. However, higher quantities don’t always guarantee higher sales, suggesting product price also plays a key role.

```python
plt.figure(figsize=(8,5))
ax = sns.scatterplot(x='PRICEEACH', y='SALES', data=df, alpha=0.6)
plt.title("Price vs Sales")
plt.show()
```

Sales tend to increase with higher product prices, but most sales are concentrated around the maximum price,
indicating premium products drive significant revenue.
